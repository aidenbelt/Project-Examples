## SEASON PREDICTION 

### Consolidate Table

#### Skater Table

```python
for col in Sskater_coef.columns:
        if col in skater_points.columns:
            Sskater_coef[f'{col}_total'] = Sskater_coef[col] * skater_points[col][0]
sum_cols = ['G_total', 'A_total', 'SHOTS_total', 'BS_total', 'SHG_total', 'SHOUTG_total']
Sskater_coef['Points'] = Sskater_coef[sum_cols].sum(axis=1)
Sskater_coef.head()
```

#### Goalie Table

```python
for col in Sgoalie_coef.columns:
        if col in goalie_points.columns:
            Sgoalie_coef[f'{col}_total'] = Sgoalie_coef[col] * goalie_points[col][0]
Gsum_cols = ['W_total', 'SV_total', 'GA_total', 'SO_total']
Sgoalie_coef['Points'] = Sgoalie_coef[Gsum_cols].sum(axis=1)
Sgoalie_coef['Pos'] = 'G'
Sgoalie_coef.head()
```

#### Combine Tables

```python
concat_cols = ['Name', 'Salary', 'Team', 'Points', 'Pos']
Splayer_coef = pd.concat([Sskater_coef[concat_cols], Sgoalie_coef[concat_cols]], axis=0)
Splayer_coef
```

### Index/Dictionary List

```python
Name_index = list(set(Splayer_coef['Name']))
Salary_dict = Splayer_coef.set_index(['Name'])['Salary'].to_dict()
Team_dict = Splayer_coef.groupby('Team')
Points_dict = Splayer_coef.set_index(['Name'])['Points'].to_dict()
Pos_dict = Splayer_coef.set_index(['Name'])['Pos'].to_dict()
```

### Define Model

```python
model = pe.ConcreteModel()
```

```python
model.x = pe.Var(Name_index, domain = pe.Binary)
```

### Maximize Points

```python
model.obj = pe.Objective(expr=sum([Points_dict[N]*model.x[N] for N in Name_index]),sense=pe.maximize)
```

### Constraints

```python
# Nine players
model.cons_9 = pe.Constraint(expr = sum(model.x[N] for N in Name_index) == 9)
# Salary Less than $50,000
model.cons_salary = pe.Constraint(expr = sum(model.x[S]*Salary_dict[S] for S in Name_index) <= 50000)
# Three different teams
model.cons_team = pe.Constraint(expr = sum(model.x[T] for T in Name_index if Team_dict['Team']) >= 3) 
# Two centers
model.cons_c = pe.Constraint(expr = sum(model.x[C] for C in Name_index if 'C' in Pos_dict[C]) >= 2)
# Three wings
model.cons_w = pe.Constraint(expr = sum(model.x[W] for W in Name_index if 'LW' or 'RW' in Pos_dict[W]) >= 3)
# Three Defense
model.cons_d = pe.Constraint(expr = sum(model.x[D] for D in Name_index if 'D' in Pos_dict[D]) >= 2)
# One goalie
model.cons_g = pe.Constraint(expr = sum(model.x[G] for G in Name_index if 'G' in Pos_dict[G]) == 1)
```

### Solve Model

```python
opt = pe.SolverFactory('glpk')
success = opt.solve(model, tee=True)
```

### Objective Value and Lineup

```python
obj_val = model.obj.expr()
game_obj_val = (obj_val // 14) 
sal_val =  model.cons_salary.body(model)

print(f'optimal objective value maximum points per game = {game_obj_val:.2f}')
print(f'optimal objective value maximum points total = {obj_val:.2f}')
print(f'optimal objective value salary = {sal_val}')
```
